```python
# coding:utf-8 

# 获取豆瓣250。
import requests 
from lxml import html 

url = 'https://movie.douban.com/top250' 
con = requests.get(url).content 
sel = html.fromstring(con) 

# 所有的信息都在class属性为info的div标签里，可以先把这个节点取出来 
for i in sel.xpath('//div[@class="info"]'):
    # 影片名称 
    title = i.xpath('div[@class="hd"]/a/span[@class="title"]/text()')[0] 
    info = i.xpath('div[@class="bd"]/p[1]/text()') 
    # 导演演员信息 
    info_1 = info[0].replace(" ", "").replace("\n", "") 
    # 上映日期 
    date = info[1].replace(" ", "").replace("\n", "").split("/")[0] 
    # 制片国家 
    country = info[1].replace(" ", "").replace("\n", "").split("/")[1] 
    # 影片类型 
    geners = info[1].replace(" ", "").replace("\n", "").split("/")[2] 
    # 评分 
    rate = i.xpath('//span[@class="rating_num"]/text()')[0] 
    # 评论人数 
    comCount = i.xpath('//div[@class="star"]/span[4]/text()')[0] 
    # 打印结果看看 
    print(title, info_1, rate, date, country, geners, comCount)
```

```python
# 获取豆瓣电影类型前一百。
# coding:utf-8 
import requests 
from lxml import html 
import time
import urllib
import json
from bs4 import BeautifulSoup
from urllib.parse import quote
import string
import xlsxwriter

# 获取tags。
def getTags():
    tags = []
    url = 'https://movie.douban.com/j/search_tags?type=movie&tag=%E7%83%AD%E9%97%A8&source='
    response = urllib.request.urlopen(url)
    result = response.read()
    # 加载json为字典
    result = json.loads(result)
    tags = result['tags']
    print(tags)
    return tags


# 获取每个tags下面的前两页电影。
def getMovies(tags):
    subjects = []
    for tag in tags:
        url = 'https://movie.douban.com/j/search_subjects?type=movie&tag='+ tag +'&page_limit=100&page_start=0'
        url = quote(url, safe=string.printable)
        response = urllib.request.urlopen(url)
        result = response.read()
        result = json.loads(result)
        olddata = result['subjects']
        for i in olddata:
            i['type'] = tag
        subjects += olddata
    print(subjects)
    return subjects

def write2excel(data,name,path):
    workbook = xlsxwriter.Workbook(path)
    worksheet = workbook.add_worksheet(name)
    
    titlerow=0
    titlecol=0
    for key in data[1].keys():
        worksheet.write(titlerow,titlecol,key)
        titlecol += 1

    row=1
    for item in data:
        col=0
        for val in item:
            worksheet.write(row,col,item[val])
            col += 1
        row += 1

    workbook.close()
    return "success"

```

```python
import pandas as pd
from pandas import DataFrame as DataFrame
from xlrd import open_workbook
import numpy as np
import re
import xlrd

import requests
from lxml import etree

import requests
from bs4 import BeautifulSoup
import traceback
import re


def get_details(excel_file):
    newexcel = xlrd.open_workbook(excel_file)
    sheet_list=open_workbook(excel_file).sheet_names()
    rettxt = ""
    movies_url_list = []
    for sheetname in sheet_list:
        sheet = newexcel.sheet_by_name(sheetname)
        df = pd.read_excel(excel_file,sheetname,usecols=[3],skiprows=[0])
        for r in df.values:
            movies_url_list.append(r[0])
    return movies_url_list


def getHTMLText(url): # 获取网页的HTML源码
    try:
        # 设置响应等待时间为30秒
        r = requests.get(url, timeout=30)
        # 检查状态码是否为200，否则抛出异常。
        r.raise_for_status()
        # 返回请求的相应内容。
        return r.text
    except:
        # 抛出异常则重试
        return getHTMLText(url)


def details2excel(list):
#     for i in list:
#         data = requests.get(i).text
#         s=etree.HTML(data)

#         Name=s.xpath('//*[@id="content"]/h1/span[1]/text()')
#         Img = s.xpath('//*[@id="mainpic"]/a/img/@src')
#         Type=s.xpath('//*[@id="info"]/span/text()')
#         Addr=s.xpath('//*[@id="info"]/text()[4]/text()')
#         releaseTime=s.xpath('//*[@id="info"]/span/text()')
#         Time=s.xpath('//*[@id="content"]/h1/span[1]/text()')
#         score=s.xpath('//*[@id="content"]/h1/span[1]/text()')
#         introduction=s.xpath('//*[@id="content"]/h1/span[1]/text()')
#         movieStartName=s.xpath('//*[@id="content"]/h1/span[1]/text()')
#         movieStartType=s.xpath('//*[@id="content"]/h1/span[1]/text()')
#         movieStartImg=s.xpath('//*[@id="content"]/h1/span[1]/text()')
    url = 'https://movie.douban.com/subject/4864908/'
    text = getHTMLText(url)
        # 将获取到的资源转换为soup对象
    soup = BeautifulSoup(text)
        # 找到所有样式为info的标签
    Name = soup.find(property="v:itemreviewed").string
    print('Name',Name)
    Img = soup.find(rel="v:image")['src']
    print('Img',Img)
    Type=soup.find(property="v:genre").string
    print('Type',Type)
    releaseTime=soup.find(property="v:initialReleaseDate").string
    print('releaseTime',releaseTime)
    Time=soup.find(property="v:runtime").string
    print('Time',Time)
    score=soup.find(property="v:average").string
    print('score',score)
    introduction=soup.find(property="v:summary").contents[0].strip()
    print('introduction',introduction)
    movieStart=soup.find_all(class_="info")
    movieStartList = []
    for idx,child in enumerate(movieStart):
        movieStartName = child.a.string
        movieStartType=child.find_all('span')[1].string
        movieStartList.append({'name':movieStartName,'type':movieStartType})
    movieStartImg=soup.find_all(class_="avatar")
    for idx,imgChild in enumerate(movieStartImg):
        startImgUrl =imgChild['style'][22:-1]
        movieStartList[idx]['imgurl'] = startImgUrl
    print('movieStartList',movieStartList)
```
